```python
import matplotlib.pyplot as plt

def generate_error_chart():
    # Dữ liệu phân tích lỗi chuẩn từ bài báo ViFactCheck
    labels = [
        'Chuỗi suy luận phức tạp\n(Complex Inferential Chain)', 
        'Ảo giác suy luận\n(Inference Hallucination)', 
        'Thất bại truy xuất\n(Evidence Retrieval Failure)', 
        'Nhập nhằng ngữ nghĩa\n(Semantic Ambiguity)', 
        'Lỗi gán nhãn\n(Annotation Error)'
    ]
    sizes = [38.0, 26.0, 21.0, 13.0, 2.0]
    colors = ['#003f5c', '#58508d', '#bc5090', '#ff6361', '#ffa600']

    # Vẽ biểu đồ tròn (Pie Chart)
    fig, ax = plt.subplots(figsize=(10, 7))
    ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 11})
    ax.axis('equal')  # Đảm bảo biểu đồ tròn đều
    
    plt.title('Phân bổ các loại lỗi dự đoán trong Xác thực Số liệu', fontsize=14, fontweight='bold', pad=20)
    plt.tight_layout()
    
    # Lưu file ảnh để chèn vào Word/Slide
    plt.savefig('error_analysis_chart.png', dpi=300)
    print("Đã xuất biểu đồ lỗi ra file 'error_analysis_chart.png' thành công!")
    plt.show()

if __name__ == "__main__":
    generate_error_chart()
```
